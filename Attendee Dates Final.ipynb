```python
import requests, json, pprint, datetime
from collections import Counter

response = requests.get('https://ct-api-challenge.herokuapp.com/')
r = response.json()
partner_dict = r['partners']
#Plan:

#Get data:
#Partner -> #Availdates
class Partner():
    def __init__(self, date, email, country):
        self.date = date
        #since we will be referring to each attendee by email
        self.name = email
        self.country = country
        self.possible_dates = []
        self.avail_dates()
    
    #a start date is only available if a partner has the day after available
    def avail_dates(self):
        
        for i in range(len(self.date)):
            self.date[i] = datetime.datetime.strptime(self.date[i], "%Y-%m-%d").date()
        for idx in range(len(self.date)):
            if self.date[idx] + datetime.timedelta(days=1) in self.date:   
                self.possible_dates.append(self.date[idx])  

#each Country object has a name, a list of emails from that country, potential dates from each partner, partners                 
class Country():
    def __init__(self, name):
        self.name = name
        self.emails = []
        self.possible_days = []
        self.partners = []
    
    #for each Country, once its partners are known, emails will be retrieved based on the partners in self.partners
    def get_emails(self):
        for person in self.partners:
            self.emails.append(person.name)
          
    
class Country_Organizer():
    partners = []
    countries = []
    conferences = []
      
    #use Partner and Country objects to fill out lists in our Organizer
    #use API locations as inputs for Partner and Country classes
    @classmethod
    def fill_lists(cls):
        for partner in partner_dict:
            single_partner = Partner(partner['availableDates'], partner['email'],partner['country'])
            cls.partners.append(single_partner)
        #prevent multiple occurences
        country_set = set()
        for single_country in partner_dict:
            country_set.add(single_country['country'])
        for c in country_set:
            country = Country(c)
            cls.countries.append(country)
    
    #add each possible conference date each time they appear
    #retrieve the most common one using Counter
    #add all the people that can attend that day
    @classmethod
    def attendance(cls):
        for person in cls.partners:
            for ctry in cls.countries:
                if person.country == ctry.name:
                    ctry.possible_days += person.possible_dates
        for ctry in cls.countries:
            ctry.meeting_date = Counter(ctry.possible_days).most_common(1)[0][0]    
        for person in cls.partners:
            for ctry in cls.countries:
                if person.country == ctry.name:
                    if ctry.meeting_date in person.possible_dates:
                        ctry.partners.append(person)
    
    #call methods and build list of dictionaries
    @classmethod
    def attendee_list(cls):
        cls.fill_lists()
        cls.attendance()
        for c in cls.countries:
            c.get_emails()
        for person in cls.partners:
            for ctry in cls.countries:
                if ctry.name == person.country:
                    dict = {
                        "Country Location" : ctry.name,
                        "Number of People Attending" : len(ctry.partners),
                        "Emails" : ctry.emails,
                        #fix the presentation of the date once printed
                        "Start Date" : str(ctry.meeting_date),
                    }
                    #no repeats
                    if dict not in cls.conferences:
                        cls.conferences.append(dict)
                
                
Country_Organizer.attendee_list()
pprint.pprint(Country_Organizer.conferences)
```
